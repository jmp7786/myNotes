```python
import plotly.offline as offline 
import plotly.graph_objs as go # jupyter notebook   

from datetime import datetime
import pandas_datareader.data as web

offline.init_notebook_mode(connected=True) 

start = datetime(2016,1,1)
end = datetime(2020,1,1)
facebook = web.DataReader('FB', 'yahoo', start, end)

trace = go.Candlestick(x=[i[0] for i in facebook['High'].items()],
                       open=[i for i in facebook['Open']], 
                       high=[i for i in facebook['High']], 
                       low=[i for i in facebook['Low']], 
                       close=[i for i in facebook['Close']]) 
data = [trace] 
layout = go.Layout(title='Facebook  ') 
fig = go.Figure(data=data, layout=layout) 
offline.iplot(fig, filename="candlestick")

```
